```python
def corr_plot(df):
    # Compute a correlation matrix and convert to long-form
    corr_mat = df.corr().stack().reset_index(name="correlation")

    # Draw each cell as a scatter point with varying size and color
    g = sns.relplot(
        data=corr_mat,
        x="level_0", y="level_1", hue="correlation", size="correlation",
        palette="vlag", hue_norm=(-1, 1), edgecolor=".7",
        height=10, sizes=(50, 350), size_norm=(-.2, .6),
    )

    # Tweak the figure to finalize
    g.set(xlabel="", ylabel="", aspect="equal")
    g.despine(left=True, bottom=True)
    g.ax.margins(.02)
    for label in g.ax.get_xticklabels():
        label.set_rotation(90)
    for artist in g.legend.legendHandles:
        artist.set_edgecolor(".7")

corr_plot(df_eda)
        
```

### Section 2. Feature Engineering

1. Monthly estimated EMI, 
2. Employment continuance in current residence
3. Estimated Start age of employment
4. Absence of Property,Savings Balance in bank

##### 1. Monthly estimated EMI
Proxy for Amount to be repaid can be got using the Amount/months. There is an EMI rate in the dataset but for simplicity lets calculate without the interest

```python
df_eda['monthly_emi']=df_eda['Principal_loan_amount']/df_eda['Months_loan_taken_for']
```

##### 2. Employment continuance in current residence
The applicant if has stayed in the same residence and is employed during this entire period he could be considered more stable and therefore less riskier. We can check this using the Years_at_current_residence and Has_been_employed_for_at_least fields.
We notice that when the emplyment period is less than the years in current residence there is higher chance of them being tagged risky.
Only exception is the group with 3 years in residence and employed with 0 and 1 year which has a higher % of applicants in non risky side. This could be because they were turning into business and not skilled or management positions.

We can add this as one of the feature


```python
pd.crosstab([df_eda['Years_at_current_residence'],df_eda['Has_been_employed_for_at_most']],
           [df_eda['high_risk_applicant']],margins=True).style.background_gradient(cmap='summer_r')
```

### Section 3. Handling Missing Values

We see that the below columns have missing values

1. Purpose - Only 1% is missing we can remove this since the volume is very minuscule
2. Property,Other EMI plans,Telephone - we can add 'Not available' as a category 
3. Savings_account_balance - we can add 'low' since most of the values are in that bucket. And create a feature to indicate rows that have missing. This would help if we are seeing any pattern in missig values 
4. Balance_in_existing_bank_account - This has a high corelation with Savings_account_balance. Therefore if Savings is high or very high or medium we can put in the 0-2 Lacs if not impute as 1 - 0 Lacs. And create a feature to indicate rows that have missing.

```python
df_feat=df_eda.copy()
df_feat.isna().sum()
```

```python
# 1. Purpose
df_feat.dropna(subset=['Purpose'],inplace=True)
```

```python
# 2. Property,Other EMI plans,Telephone
cols=['Property','Telephone','Other_EMI_plans']
for col in cols:
    df_feat.loc[df_feat[col].isna(),col]='Not Available'
```

```python
#3. Savings_account_balance 
df_feat['Savings_balance_not_available']=0
df_feat.loc[df_feat['Savings_account_balance'].isna(),'Savings_balance_not_available']=1

df_feat.loc[df_feat['Savings_account_balance'].isna(),'Savings_account_balance']='Low'
```

```python
# 4. Balance_in_existing_bank_account
# Adding variable for NAs
df_feat['Balance_not_available']=0
df_feat.loc[df_feat['Balance_in_existing_bank_account_(upper_limit_of_bucket)'].isna(),'Balance_not_available']=1
```

```python
#imputing Balance_in_existing_bank_account for high, medium balance accounts
df_feat.loc[df_feat['Balance_in_existing_bank_account_(lower_limit_of_bucket)'].isna()&
            df_feat['Savings_account_balance'].isin(['High', 'Very high', 'Medium']),
            'Balance_in_existing_bank_account_(lower_limit_of_bucket)']='0'
                                                    
df_feat.loc[df_feat['Balance_in_existing_bank_account_(upper_limit_of_bucket)'].isna()&
            df_feat['Savings_account_balance'].isin(['High', 'Very high', 'Medium']),
            'Balance_in_existing_bank_account_(upper_limit_of_bucket)']='2 lac'
```

```python
#imputing Balance_in_existing_bank_account for low savings bacnk accounts
df_feat.loc[df_feat['Balance_in_existing_bank_account_(lower_limit_of_bucket)'].isna(),
            'Balance_in_existing_bank_account_(lower_limit_of_bucket)']='<0'
                                                    
df_feat.loc[df_feat['Balance_in_existing_bank_account_(upper_limit_of_bucket)'].isna(),
            'Balance_in_existing_bank_account_(upper_limit_of_bucket)']='0'
```

### Section 4.  Evaluation Metric

For the model we build, we need to identify the high risk applications based on the application details. One of the key considerations is that, it is not okay to call a high risk applicant as non high risk. But calling a non high risk as high risk is still acceptable. This is to ensure that the prinicpal amount in loans are not getting written off going behind for a fraction of interest

In terms of the metrics, we need to more focused on 
1. Recall - predicting almost all the True positives correctly as High risk than focusing on the Precision.

2. Wieghted Accuracy - As per the cost we can estimate the weighted Accuarcy score by adding wieghts as per the cost matrix shared in the document.


### Section 5. Baseline Model - Non ML

We can have a baseline model which predicts the majority class at all times. In this case since majority is non high risk which is 70% our baseline model will not predict high risk at all. 

Therefore lets predict a random 50% of them as either of the class



```python
from sklearn.metrics import confusion_matrix,precision_score,recall_score,accuracy_score
```

```python
y_true=df_eda['high_risk_applicant']
y_pred=np.random.randint(2,size=1000,)
```

```python
cm = confusion_matrix(y_pred=y_pred,y_true=y_true)
print(cm)
```

```python
precision_score(y_pred=y_pred,y_true=y_true)
```

```python
recall_score(y_pred=y_pred,y_true=y_true)
```

```python
TN=cm[0,0]
TP=cm[1,1]
FN=cm[1,0]
FP=cm[0,1]
```

```python
Accuracy_with_cost = (TN+TP)/(TN+TP+(4*FN)+FP)
print(Accuracy_with_cost)
```
